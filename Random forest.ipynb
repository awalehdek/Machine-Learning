```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Load Iris dataset
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = pd.Series(iris.target)



X.head()
```

```python
y.head()
```

#LabelEncoder from the sklearn.preprocessing module is used to convert categorical labels into numerical values to easier for machine learning algorithms

```python
# Encode target variable
le = LabelEncoder()
y_encoded = le.fit_transform(y)

y_encoded
```

```python

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Train Random Forest model
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

```

```python
# Make predictions
y_pred = rf_model.predict(X_test)
```

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

#Heatmap
sns.heatmap: is used to create a heatmap of the confusion matrix.

annot=True ensures that the values are annotated on the heatmap.

fmt='d' specifies that the annotation should be displayed as integers.

cmap='Blues' sets the color map to blues.

xticklabels and yticklabels are set to the names of the target classes


```python
from sklearn.metrics import confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# Compute the confusion matrix
cm = confusion_matrix(y_test, y_pred)
# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
